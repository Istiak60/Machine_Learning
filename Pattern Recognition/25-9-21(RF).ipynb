```python
import io
```

```python
import pandas as pd
```

```python
df= pd.read_csv(io.BytesIO(upload['cardio_train.csv']), sep=';')
```

```python
df.head()
```

```python
x=df.drop(['cardio'],axis=1)
y=df['cardio']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,train_size=0.75, test_size=0.25, random_state=50)
```

```python
#feature Selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2
```

```python
feature= SelectKBest(score_func=f_classif,k='all')
feature.fit(x,y)

```

```python
feature
```

```python
pd.DataFrame(feature.scores_)
```

```python
#import naivebayes
from sklearn.naive_bayes import GaussianNB
NB= GaussianNB()
```

```python
NB.fit(x_train,y_train)
NB.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
DT = RandomForestClassifier()
DT.fit(x_train,y_train)
```

```python
DT.score(x_test, y_test)
```
